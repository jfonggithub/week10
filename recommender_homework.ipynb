```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]

    
```

```python
newTable = pd.DataFrame(index=df.columns, columns=['Pearson with G. of the Galaxy', 'Pearson with Divergent', 'Predicted rating'])
newTable['Pearson with G. of the Galaxy'] = 0
newTable['Pearson with Divergent'] = 0
newTable['Predicted rating'] = 0
```

```python
# It appears to be difficult to predict the rating based on my ratings on the movies as 
# the pearson scores doesn't have a consistent pattern among my ratings.
# Below is a primitive way of predicting the rating.
def populatePearsonScores(data, table):
    for index, row in table.iterrows():
            table.loc[index, 'Pearson with G. of the Galaxy'] = pearson(data[index], data['Guardians of the Galaxy'])
            table.loc[index, 'Pearson with Divergent'] = pearson(data[index], data['Divergent'])
            if row['Pearson with G. of the Galaxy'] > 0.68:
                table.loc[index, 'Predicted rating'] = 2
            elif row['Pearson with G. of the Galaxy'] == 0.0:
                table.loc[index, 'Predicted rating'] = 5
            else:
                table.loc[index, 'Predicted rating'] = 4
```

```python
populatePearsonScores(df, newTable)
```
